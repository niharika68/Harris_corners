```python
def rgb2grayscale(im):
        if  len(im.shape) > 2:
            if im.shape[2] == 3: # Convert RGB image to Grayscale
                r, g, b = im[:, :, 0], im[:, :, 1], im[:, :, 2]
                grayscale = 0.2989 * r + 0.5870 * g + 0.1140 * b
                return grayscale
        else:
            return im
```

```python
from PIL import Image
import numpy as np
import math
import scipy.misc


def sobel_convolute(img,kernel_common):
    rows,cols=img.shape
    
    op_image=np.zeros([rows,cols])
    for i in np.arange(1,rows-1):
         for j in np.arange(1,cols-1):
                op_image[i,j]=np.sum(kernel_common*img[i-1:i+2,j-1:j+2])
                
    return op_image

kernel_x=np.array(([-1, 0, 1],[-2, 0, 2],[-1, 0, 1]), dtype='int')
kernel_y=np.array(([-1, -2, -1],[0, 0, 0],[1, 2, 1]),dtype='int')
#img = Image.open('apple.jpg')
img = Image.open('test5.jpg')
img2array = rgb2grayscale(np.array(img)) 

#step 1 find the gradient 
Ix=sobel_convolute(img2array,kernel_x)
Iy=sobel_convolute(img2array,kernel_y)
```

```python
# step 2 find Ixx, Iyy and Ixy
Ixx=Ix**2
Iyy = Iy**2
Ixy = Iyy*Ixx

```

```python
#calculate sum of all the 
k = 0.004
window_size = 6
offset = int(window_size/2)
height, width = img2array.shape
harris_response =[]

for y in range(offset, height-offset):
   
    for x in range(offset, width-offset):

        Sxx = np.sum(Ixx[y-offset:y+1+offset, x-offset:x+1+offset])
        Syy = np.sum(Iyy[y-offset:y+1+offset, x-offset:x+1+offset])
        Sxy = np.sum(Ixy[y-offset:y+1+offset, x-offset:x+1+offset])
        det = (Sxx * Syy) - (Sxy**2)
        trace = Sxx + Syy
        r = det - k*(trace**2)
        harris_response.append([x,y,r])
```

```python
#storing corner in the harris corner 
newImg = np.array(img)
#color_img = cv2.cvtColor(newImg, cv2.COLOR_GRAY2RGB)

for response in harris_response:
    x, y, r = response
    if r > 0:
        print(r)
        newImg[y,x,2] = 255
        newImg[y,x,0] = 0
        newImg[y,x,1] = 0
```

```python
disp = Image.fromarray(newImg)
```
